# Importing Libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import sklearn.datasets
%matplotlib inline
print("Setup Completed")
```

# Loading Data

```python
breast_cancer = sklearn.datasets.load_breast_cancer()
print(breast_cancer)
```

```python
X = breast_cancer.data
y = breast_cancer.target
```

# Importing Data Into Pandas DataFrame

```python
cancer = pd.DataFrame(X,columns = breast_cancer.feature_names)
```

```python
cancer["class"] = y
```

```python
cancer.head(2)
```

# Data Exploration

```python
cancer.describe()
```

```python
cancer.isnull().sum()
```

```python
cancer.corr()
```

```python
sns.heatmap(cancer.corr(),linecolor='green',linewidth = 0.2,cmap='viridis')
```

```python
cancer['class'].nunique()
```

# Splitting Data

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(cancer.drop('class',axis=1), cancer['class'], test_size=0.33, random_state=42)
```

# Random Forest Model 

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
rtree = RandomForestClassifier(n_estimators=2000)
```

```python
rtree.fit(X_train,y_train)
```

# Predictions

```python
predictions = rtree.predict(X_test)
```

# Evaluation 

```python
from sklearn.metrics import confusion_matrix,classification_report,r2_score
```

```python
print("Confusion Metrics: ",confusion_matrix(y_test,predictions))
print("Classification Report: ",classification_report(y_test,predictions))
print("Score: ",r2_score(y_test,predictions))
```

## Artificial Neural Networks

```python
from sklearn.preprocessing import StandardScaler
```

```python
ss = StandardScaler()
```

```python
X_train = ss.fit_transform(X_train)
```

```python
X_test = ss.fit_transform(X_test)
```

# Adding Layers

```python
import keras
from keras.models import Sequential
from keras.layers import Dense, Dropout,Input
```

```python
classifier = Sequential()
```

```python
X_train.shape
```

```python
classifier.add(Input(shape=(30,)))
classifier.add(Dense(16,activation='relu'))
classifier.add(Dropout(0.1))
classifier.add(Dense(16,activation='relu'))
classifier.add(Dropout(0.1))
classifier.add(Dense(1,activation='sigmoid'))

# model.add(tf.keras.Input(shape=(381,30)))
# model.add(tf.keras.layers.Dense(16, activation='relu'))
# Now the model will take as input arrays of shape (None, 16)
# and output arrays of shape (None, 32).
```

```python
classifier.compile(optimizer='adam',loss='binary_crossentropy',metrics =['accuracy'])
```

```python
classifier.fit(X_train,y_train, batch_size=100,epochs=150)
```

```python
n_pred = classifier.predict(X_test)
```

```python
n_pred = n_pred.round(0)
```

# Evaluation

```python
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y_test, n_pred)
cm
```

```python
print(classification_report(y_test,n_pred))
```

```python
print("Our accuracy is {}%".format(((cm[0][0] + cm[1][1])/188)*100))
```
